```python

import logging

# Parameters
MODELNAME = "test"  # Replace with your model's name
MODELEPOCH = 9600  # Replace with your model's epoch number

# Setup logging
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(message)s")

# Define directories
BASE_DIR = os.path.abspath("Retrieval-based-Voice-Conversion-WebUI")  # Path to your project folder
LOGS_DIR = os.path.join(BASE_DIR, "logs", MODELNAME)
WEIGHTS_DIR = os.path.join(BASE_DIR, "weights")
BACKUP_DIR = os.path.expanduser("~/Documents/Model_Backups")  # Desired backup folder

# Ensure required directories exist
os.makedirs(LOGS_DIR, exist_ok=True)
os.makedirs(WEIGHTS_DIR, exist_ok=True)
os.makedirs(BACKUP_DIR, exist_ok=True)

logging.info(f"Directories ensured: {LOGS_DIR}, {WEIGHTS_DIR}, {BACKUP_DIR}")

# Simulate backup process (if files exist)
try:
    # Backup generator and discriminator model files
    logging.info(f"Attempting to backup Generator and Discriminator files for epoch {MODELEPOCH}.")
    shutil.copy(os.path.join(LOGS_DIR, f"G_{MODELEPOCH}.pth"), os.path.join(BACKUP_DIR, f"{MODELNAME}_G_{MODELEPOCH}.pth"))
    shutil.copy(os.path.join(LOGS_DIR, f"D_{MODELEPOCH}.pth"), os.path.join(BACKUP_DIR, f"{MODELNAME}_D_{MODELEPOCH}.pth"))
except FileNotFoundError:
    logging.warning(f"Generator or Discriminator files for epoch {MODELEPOCH} not found. Skipping file backup.")

# Backup additional index and numpy files
logging.info("Checking for additional index and numpy files to backup.")
for file in os.listdir(LOGS_DIR):
    if file.startswith("added_") and file.endswith(".index") or file.startswith("total_") and file.endswith(".npy"):
        shutil.copy(os.path.join(LOGS_DIR, file), BACKUP_DIR)

# Attempt to backup weight file
try:
    logging.info(f"Attempting to backup weight file: {MODELNAME}.pth.")
    shutil.copy(os.path.join(WEIGHTS_DIR, f"{MODELNAME}.pth"), os.path.join(BACKUP_DIR, f"{MODELNAME}{MODELEPOCH}.pth"))
except FileNotFoundError:
    logging.warning(f"Weight file {MODELNAME}.pth not found. Skipping file backup.")

logging.info(f"Backup process completed. Files saved in: {BACKUP_DIR}")
```
